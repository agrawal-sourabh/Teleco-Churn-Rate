```python
#Importing Packages
import numpy as np 
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

```python
df.dtypes
```

```python
for item in df.columns:
    print(item)
    print (df[item].unique())
```

```python
#Converting Strings to Lowercase [Fields with integer would raise an exception] 
for item in df.columns:
    try:
        df[item] = df[item].str.lower()
    except:
        print(item, "couldn't convert")
df.head()
```

```python
churn_in_last_month = df[df['Churn'] == 'yes']
churn_in_last_month
```

```python
for item in df.columns:
    print(item)
    print(churn_in_last_month[item].value_counts())
```
